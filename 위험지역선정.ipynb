# 범주형변수 groupby sum 중에 max 값은 1, 나머지는 0

```python
# 그룹별 열을 식별하여 처리
def set_max_to_one(df3, prefix):
    # prefix로 시작하는 열을 필터링
    group_cols = [col for col in df3.columns if col.startswith(prefix)]
    # 최대값을 가지는 열을 1로 설정하고 나머지는 0으로 설정
    df3[group_cols] = (df3[group_cols].eq(df3[group_cols].max(axis=1), axis=0)).astype(int)

# 그룹별로 최대값 처리
prefixes = ['요일_', '법규위반_', '노면상태_', '기상상태_', '사고유형_1', '가해운전자 연령대_', '피해운전자 연령대_','사고시간대_', '도로형태_']
for prefix in prefixes:
    set_max_to_one(df3, prefix)
```

# 연속형변수들 groupby

```python
df4 = df1[['시군구_구동', 'EPDO', 'cctv_count', 'bus_count', '주차장_count',
       '어린이보호구역_count']]
# 그룹별 평균값 계산
mean_values = df4.groupby('시군구_구동').mean()

# 그룹별 항목 수 계산
count_values = df4.groupby('시군구_구동').size()

# 그룹별 epdo 최대값 계산
max_epdo_values = df4.groupby('시군구_구동')['EPDO'].max()

# 두 결과를 하나의 데이터프레임으로 합치기
result = mean_values.assign(Count=count_values, epdo_max=max_epdo_values)

result.rename(columns={'EPDO' : '실제_EPDO',
                       'Count' : '사고건수'}, inplace=True)
result
```
